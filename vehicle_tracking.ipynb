```python
import cv2
import os
from ultralytics import YOLO


def detect_objects(input_path, output_folder):
    """
    Detect objects in an image or video and save frames with bounding boxes and labels.
    :param input_path: path to image or video
    :param output_folder: folder where frames will be saved
    """

    # Load YOLO model
    model = YOLO("yolov8n.pt")  # uses COCO pretrained model

    # Create output folder if not exists
    os.makedirs(output_folder, exist_ok=True)

    # Check if input is image
    if input_path.lower().endswith((".jpg", ".jpeg", ".png")):
        img = cv2.imread(input_path)
        results = model(img)

        # Draw bounding boxes
        annotated = results[0].plot()

        # Save annotated image
        save_path = os.path.join(output_folder, "output_image.jpg")
        cv2.imwrite(save_path, annotated)
        print(f"âœ… Image processed and saved at: {save_path}")

    else:
        # Process video
        cap = cv2.VideoCapture(input_path)
        frame_count = 0

        if not cap.isOpened():
            print("âŒ Error: Could not open video.")
            return

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            results = model(frame)
            annotated = results[0].plot()

            frame_count += 1
            save_path = os.path.join(output_folder, f"frame_{frame_count}.jpg")
            cv2.imwrite(save_path, annotated)

        cap.release()
        print(f"ğŸ¥ Video processed. {frame_count} frames saved to: {output_folder}")


def main():
    input_path = "/content/test2.jpg"
    output_folder = "test1"

    detect_objects(input_path, output_folder)


if __name__ == "__main__":
    main()
```
